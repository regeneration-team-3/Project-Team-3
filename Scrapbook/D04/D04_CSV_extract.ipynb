# Creating CSV for Power BI

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("extracted.data.csv")
```

```python
# extract car brands from car name
def getBrand(x):
    brand = x.lower().split()[0]
    return brand
df['brand'] = df['car name'].map(getBrand)
```

```python
# δημιουργία bag of words από brands με στόχο την εύρεση τυπογραφικών
bagOfWords = {}
for name in df['brand']:
    words = name.lower().replace('(', '').replace(')', '').split()
    for word in words:
        if word not in bagOfWords:
            bagOfWords[word] = 0
        bagOfWords[word] += 1
```

```python
def getBrand(x):
    brand = x.lower().split()[0]
    if (brand == 'vw' or brand == 'vokswagen'):
        return 'volkswagen'
    if (brand == 'toyouta'):
        return 'toyota'
    if (brand == 'maxda'):
        return 'mazda'
    if (brand == 'chevroelt'):
        return 'chevrolet'
    if (brand == 'mercedes-benz'):
        return 'mercedes'
    return brand
df['brand'] = df['car name'].map(getBrand)
```
